### Plots
- Code 1

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 20, 100)
y = 2*np.sin(x)
y1 = 2*np.cos(x)
fig, ax = plt.subplots()
ax.plot(x, y, linewidth=2.0, color='r')
ax.plot(x, y1, linewidth=2.0, color='g')
ax.plot(x, x**(1/2), linewidth=2.0, color='b')
ax.set(xlim=(0, 20),   ylim=(-3, 3))
ax.set_title("Plotting of Curves")
plt.show()
```

### Sub-Plots
- Multiple Graph in the same output window
- Method 1

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,10.0,0.1)
fig , ax = plt.subplots(2,3)
ax[0,0].plot(x,x)
ax[0,1].plot(x,x**2)
ax[0,2].plot(x,np.sin(x))
ax[1,0].plot(x,np.sqrt(x))
ax[1,1].plot(x,1/x)
ax[1,2].plot(x,np.cos(x))
plt.show()
# plt.savefig('demo1.png')
plt.close()
```

- Method 2

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 20, 100)
y = 2*np.sin(x)
y1= 2*np.cos(x)
y2= np.cos(2*x)
y3= np.sin(2*x)
fig, axs = plt.subplots(2, 2, layout='constrained')
pc = axs[0, 0].plot(x, y, linewidth=1.0, color='r')
axs[0, 0].set_title('2sin(x)')
co = axs[0, 1].plot(x,y1, linewidth=1.0, color='g')
axs[0, 1].set_title('2cos(x)')
pc1 = axs[1, 0].plot(x, y2, linewidth=1.0, color='b')
axs[1, 0].set_title('cos(2x)')
co1 = axs[1, 1].plot(x,y3, linewidth=1.0, color='m')
axs[1, 1].set_title('sin(2x)')
plt.show()
```

### Surface Plotting

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
# print(x)
y = x.copy().T # transpose
z = np.sqrt(x **2 + y** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('Surface plot z=sqrt(x^2+y^2)')

plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T # transpose
z = np.cos((x*y)/(1+2*(y**2)))

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z,cmap='RdPu', edgecolor='none')
ax.set_title('Surface plot z=cos((x*y)/(1+2*(y^2))))')

plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T # transpose
z = 1 - x**2 - y**2

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z,cmap='bone', edgecolor='none')
ax.set_title('Surface plot z=sin(x^2+y^2)')

plt.show()
```

### Surface Subplotting

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1,2,1,projection = '3d')
x = np.outer(np.linspace(-2,2,30),np.ones(30))
y = x.copy().T
z = (x * y**6 + y)
ax.plot_surface(x,y,z,cmap = "copper",edgecolor = "none")
ax.set_title("Surface plot z = x(y^6) + y")
ax = fig.add_subplot(1,2,2,projection = "3d")
z2 = np.sin(2*x)
ax.plot_surface(x,y,z2,cmap = "seismic",edgecolor = "none")
ax.set_title("Surface plot z2 = sin(2x)")
plt.show()
```

- Four 3D Subplots in one window

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()

x = np.outer(np.linspace(-2,2,30),np.ones(30))
y = x.copy().T

z = np.cos(x*y)
ax = fig.add_subplot(2,2,1,projection = '3d')
ax.plot_surface(x,y,z,cmap = "copper",edgecolor = "none", rstride=2, cstride=2, antialiased=False)
ax.set_title("Surface plot z = cos(xy)")
# surf = ax.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap=cm.coolwarm,linewidth=0, antialiased=False)

ax = fig.add_subplot(2,2,2,projection = '3d') 
z = np.sin(x*y)
ax.plot_surface(x,y,z,cmap = "copper",edgecolor = "none", rstride=2, cstride=2, antialiased=False)
ax.set_title("Surface plot z = sin(xy)")

ax = fig.add_subplot(2,2,3,projection = '3d')
z = np.cos(x+y)
ax.plot_surface(x,y,z,cmap = "copper",edgecolor = "none", rstride=2, cstride=2, antialiased=False)
ax.set_title("Surface plot z = cos(x+y)")

ax = fig.add_subplot(2,2,4,projection = '3d')
z = np.sin(x+y)
ax.plot_surface(x,y,z,cmap = "copper",edgecolor = "none", rstride=2, cstride=2, antialiased=False)
ax.set_title("Surface plot z = sin(x+y)")

plt.show()
```

### Parametric Curve

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['legend.fontsize'] = 10
fig = plt.figure()
ax = fig.gca(projection='3d')
# ax = fig.gca()
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = 2 # radius
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()
plt.show()
```
